# 🫁 Medical Captioning — Détection d'anomalies sur radiographies thoraciques

---

## 🎯 Objectif du projet

Ce projet construit un **pipeline d'analyse automatique de radiographies thoraciques** en deux étapes :

1. **Détection des anomalies** : un modèle YOLOv8 détecte et localise les pathologies pulmonaires avec un indice de confiance pour chaque détection.
2. **Génération d'un rapport clinique** : les détections sont transmises à MedGemma-4B (Google) qui produit un rapport structuré décrivant les anomalies et des recommandations.

---

## 📦 Dataset — VinBigData

15 000 radiographies thoraciques annotées par des radiologues, 15 classes de pathologies au format YOLO avec WBF (Weighted Box Fusion).

|---|---|
| Aortic enlargement | Atelectasis |
| Calcification | Cardiomegaly |
| Consolidation | ILD |
| Infiltration | Lung Opacity |
| Nodule/Mass | Other lesion |
| Pleural effusion | Pleural thickening |
| Pneumothorax | Pulmonary fibrosis |
| No finding | |

Source : `buithanhxuan/vinbigdata-yolo-dataset-with-wbf-3x-downscaled` (Kaggle)

### ⚖️ Création du subset équilibré — 5000 images (70% train / 20% val / 10% test)

```python
import os
import shutil
import random
from collections import defaultdict, Counter

random.seed(42)

base_dir2 = '/content/vinbigdata-yolo/vinbigdata-yolo-dataset-with-wbf-3x-downscaled'
subset_dir = '/content/vinbigdata-yolo-subset'

# ── PARAMÈTRES ─────────────────────────────────────────
TOTAL_IMAGES  = 5000
TRAIN_SIZE    = int(TOTAL_IMAGES * 0.70)  # 3500
VAL_SIZE      = int(TOTAL_IMAGES * 0.20)  # 1000
TEST_SIZE     = int(TOTAL_IMAGES * 0.10)  #  500
MAX_CLASS_14  = 1500  # plafond spécifique pour la classe dominante
MAX_OTHER     = 1200  # plafond pour les autres classes

# ── CRÉE LES DOSSIERS ──────────────────────────────────
for split in ['train', 'val', 'test']:
    os.makedirs(f'{subset_dir}/{split}/images', exist_ok=True)
    os.makedirs(f'{subset_dir}/{split}/labels', exist_ok=True)

# ── COLLECTE TOUTES LES IMAGES ─────────────────────────
all_data = []
for split in ['train', 'val', 'test']:
    img_dir = os.path.join(base_dir2, split, 'images')
    lbl_dir = os.path.join(base_dir2, split, 'labels')
    for fname in os.listdir(lbl_dir):
        if not fname.endswith('.txt'):
            continue
        img_name = fname.replace('.txt', '.jpg')
        img_path = os.path.join(img_dir, img_name)
        lbl_path = os.path.join(lbl_dir, fname)
        if os.path.exists(img_path):
            all_data.append((img_path, lbl_path))

random.shuffle(all_data)
print(f"Total images disponibles : {len(all_data)}")

# ── SÉLECTION ÉQUILIBRÉE ───────────────────────────────
selected = []
selected_paths = set()
class_counts = defaultdict(int)

for img_path, lbl_path in all_data:
    if len(selected) >= TOTAL_IMAGES:
        break

    with open(lbl_path, 'r') as f:
        lines = [l.strip() for l in f if l.strip()]

    classes_in_image = [int(l.split()[0]) for l in lines]

    # Vérifie les plafonds
    skip = False
    for c in set(classes_in_image):
        cap = MAX_CLASS_14 if c == 14 else MAX_OTHER
        if class_counts[c] >= cap:
            skip = True
            break

    if skip:
        continue

    selected.append((img_path, lbl_path))
    selected_paths.add(img_path)
    for c in classes_in_image:
        class_counts[c] += 1

print(f"Images sélectionnées : {len(selected)}")

# ── Si on n'atteint pas 5000, on complète sans restriction ──
if len(selected) < TOTAL_IMAGES:
    print(f"⚠️ Seulement {len(selected)} images avec plafond, on complète sans restriction...")
    for img_path, lbl_path in all_data:
        if len(selected) >= TOTAL_IMAGES:
            break
        if img_path not in selected_paths:
            selected.append((img_path, lbl_path))
            selected_paths.add(img_path)

print(f"Images finales : {len(selected)}")

# ── SPLIT TRAIN / VAL / TEST ───────────────────────────
random.shuffle(selected)
train_set = selected[:TRAIN_SIZE]
val_set   = selected[TRAIN_SIZE:TRAIN_SIZE + VAL_SIZE]
test_set  = selected[TRAIN_SIZE + VAL_SIZE:TRAIN_SIZE + VAL_SIZE + TEST_SIZE]

print(f"Train: {len(train_set)} | Val: {len(val_set)} | Test: {len(test_set)}")

# ── COPIE LES FICHIERS ─────────────────────────────────
def copy_split(data, split):
    for img_path, lbl_path in data:
        img_name = os.path.basename(img_path)
        lbl_name = os.path.basename(lbl_path)
        shutil.copy(img_path, f'{subset_dir}/{split}/images/{img_name}')
        shutil.copy(lbl_path, f'{subset_dir}/{split}/labels/{lbl_name}')

copy_split(train_set, 'train')
copy_split(val_set,   'val')
copy_split(test_set,  'test')

print("✅ Subset créé avec succès !")

# ── VÉRIFICATION DISTRIBUTION FINALE ──────────────────
for split in ['train', 'val', 'test']:
    lbl_dir = f'{subset_dir}/{split}/labels'
    counts = Counter()
    for fname in os.listdir(lbl_dir):
        with open(os.path.join(lbl_dir, fname), 'r') as f:
            for line in f:
                if line.strip():
                    counts[int(line.split()[0])] += 1
    print(f"\nDistribution {split}: {dict(sorted(counts.items()))}")
```

### 🖼️ Vérification visuelle — Bounding boxes sur images aléatoires

```python
import cv2
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import random
import os

random.seed(99)

subset_dir = '/content/vinbigdata-yolo-subset'

# Noms des classes VinBigData
CLASS_NAMES = {
    0: 'Aortic enlargement', 1: 'Atelectasis', 2: 'Calcification',
    3: 'Cardiomegaly',       4: 'Consolidation', 5: 'ILD',
    6: 'Infiltration',       7: 'Lung Opacity',  8: 'Nodule/Mass',
    9: 'Other lesion',      10: 'Pleural effusion', 11: 'Pleural thickening',
    12: 'Pneumothorax',     13: 'Pulmonary fibrosis', 14: 'No finding'
}

COLORS = plt.cm.get_cmap('tab20', 15)

def draw_boxes(ax, img_path, lbl_path, title):
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    h, w = img.shape[:2]
    ax.imshow(img)
    ax.set_title(title, fontsize=8, fontweight='bold')
    ax.axis('off')

    with open(lbl_path, 'r') as f:
        for line in f:
            parts = line.strip().split()
            if len(parts) < 5:
                continue
            c, cx, cy, bw, bh = int(parts[0]), float(parts[1]), float(parts[2]), float(parts[3]), float(parts[4])
            x1 = (cx - bw/2) * w
            y1 = (cy - bh/2) * h
            bw_px = bw * w
            bh_px = bh * h
            color = COLORS(c)
            rect = patches.Rectangle((x1, y1), bw_px, bh_px,
                                      linewidth=1.5, edgecolor=color, facecolor='none')
            ax.add_patch(rect)
            ax.text(x1, y1 - 4, CLASS_NAMES.get(c, str(c)),
                    color='white', fontsize=6, backgroundcolor=color,
                    bbox=dict(facecolor=color, alpha=0.7, pad=1, edgecolor='none'))

# ── SÉLECTIONNE 3 IMAGES PAR SPLIT ────────────────────
fig, axes = plt.subplots(3, 3, figsize=(16, 14))
fig.suptitle('Vérification du subset — Bounding Boxes', fontsize=14, fontweight='bold')

for row, split in enumerate(['train', 'val', 'test']):
    img_dir = f'{subset_dir}/{split}/images'
    lbl_dir = f'{subset_dir}/{split}/labels'
    files = [f for f in os.listdir(lbl_dir) if f.endswith('.txt')]
    samples = random.sample(files, 3)

    for col, fname in enumerate(samples):
        img_name = fname.replace('.txt', '.jpg')
        img_path = os.path.join(img_dir, img_name)
        lbl_path = os.path.join(lbl_dir, fname)
        draw_boxes(axes[row][col], img_path, lbl_path, f'{split.upper()} — {fname[:15]}...')

plt.tight_layout()
plt.savefig('/content/verification_subset.png', dpi=150, bbox_inches='tight')
plt.show()
print("✅ Vérification affichée !")
```

### ⚖️ Création du subset étendu — 7500 images avec priorité classes rares

```python
import os, shutil, random
from collections import defaultdict, Counter

random.seed(42)

base_dir2 = '/content/vinbigdata-yolo/vinbigdata-yolo-dataset-with-wbf-3x-downscaled'
subset_dir     = '/content/vinbigdata-yolo-subset'
subset_dir_new = '/content/vinbigdata-yolo-subset-7500'

# Images déjà sélectionnées
already_selected = set()
for split in ['train', 'val', 'test']:
    img_dir = f'{subset_dir}/{split}/images'
    for fname in os.listdir(img_dir):
        already_selected.add(fname)
print(f"Images déjà sélectionnées : {len(already_selected)}")

# Classes prioritaires (les plus rares)
PRIORITY_CLASSES = {1, 12, 4, 2, 6}
TARGET_NEW = 2500

# Collecte les images restantes
remaining = []
for split in ['train', 'val', 'test']:
    img_dir = os.path.join(base_dir2, split, 'images')
    lbl_dir = os.path.join(base_dir2, split, 'labels')
    for fname in os.listdir(lbl_dir):
        if not fname.endswith('.txt'):
            continue
        img_name = fname.replace('.txt', '.jpg')
        if img_name in already_selected:
            continue
        img_path = os.path.join(img_dir, img_name)
        lbl_path = os.path.join(lbl_dir, fname)
        if os.path.exists(img_path):
            remaining.append((img_path, lbl_path))

random.shuffle(remaining)
print(f"Images disponibles restantes : {len(remaining)}")

# Passe 1 — priorité classes rares
selected_new = []
selected_new_paths = set()

for img_path, lbl_path in remaining:
    if len(selected_new) >= TARGET_NEW:
        break
    with open(lbl_path, 'r') as f:
        classes = set(int(l.split()[0]) for l in f if l.strip())
    if classes & PRIORITY_CLASSES:
        selected_new.append((img_path, lbl_path))
        selected_new_paths.add(img_path)

print(f"Après passe prioritaire : {len(selected_new)} images")

# Passe 2 — compléter si pas assez
for img_path, lbl_path in remaining:
    if len(selected_new) >= TARGET_NEW:
        break
    if img_path not in selected_new_paths:
        selected_new.append((img_path, lbl_path))
        selected_new_paths.add(img_path)

print(f"Total nouvelles images : {len(selected_new)}")

# Copie le subset existant
for split in ['train', 'val', 'test']:
    shutil.copytree(
        f'{subset_dir}/{split}',
        f'{subset_dir_new}/{split}',
        dirs_exist_ok=True
    )
print("✅ Subset 5000 copié !")

# Répartit les 2500 nouvelles en 70/20/10
random.shuffle(selected_new)
n = len(selected_new)
new_train = selected_new[:int(n*0.70)]
new_val   = selected_new[int(n*0.70):int(n*0.90)]
new_test  = selected_new[int(n*0.90):]

def copy_split(data, split):
    for img_path, lbl_path in data:
        shutil.copy(img_path, f'{subset_dir_new}/{split}/images/{os.path.basename(img_path)}')
        shutil.copy(lbl_path, f'{subset_dir_new}/{split}/labels/{os.path.basename(lbl_path)}')

copy_split(new_train, 'train')
copy_split(new_val,   'val')
copy_split(new_test,  'test')

CLASS_NAMES = {
    0:'Aortic_enlargement', 1:'Atelectasis', 2:'Calcification',
    3:'Cardiomegaly', 4:'Consolidation', 5:'ILD', 6:'Infiltration',
    7:'Lung_Opacity', 8:'Nodule/Mass', 9:'Other_lesion',
    10:'Pleural_effusion', 11:'Pleural_thickening', 12:'Pneumothorax',
    13:'Pulmonary_fibrosis', 14:'No finding'
}

print("\n" + "="*60)
print("DISTRIBUTION FINALE DU SUBSET 7500")
print("="*60)

total_counts = Counter()
for split in ['train', 'val', 'test']:
    lbl_dir = f'{subset_dir_new}/{split}/labels'
    counts = Counter()
    for fname in os.listdir(lbl_dir):
        with open(os.path.join(lbl_dir, fname), 'r') as f:
            for line in f:
                if line.strip():
                    counts[int(line.split()[0])] += 1
    total_counts += counts

    n_images = len(os.listdir(f'{subset_dir_new}/{split}/images'))
    print(f"\n── {split.upper()} ({n_images} images) ──")
    for c in range(15):
        bar = '█' * (counts[c] // 100)
        print(f"  {CLASS_NAMES[c]:25s} : {counts[c]:5d}  {bar}")

print(f"\n── TOTAL ──")
for c in range(15):
    bar = '█' * (total_counts[c] // 200)
    print(f"  {CLASS_NAMES[c]:25s} : {total_counts[c]:5d}  {bar}")
```

# 🧠 PARTIE 4 — Pipeline LLM : Rapports cliniques avec MedGemma

**Objectif** : connecter YOLO à un LLM médical pour générer des rapports cliniques automatiques.

**Pipeline complet** :
1. YOLOv8m détecte les anomalies
2. Chaque détection reçoit un indice de confiance : 🟢 Haute (>70%) / 🟡 Moyenne (45-70%) / 🔴 Faible (<45%)
3. Un prompt structuré est construit avec les détections et leurs niveaux de confiance
4. **MedGemma-4B-IT** analyse l'image annotée et génère un rapport clinique

**Modèle LLM** : `google/medgemma-4b-it`
- 4B paramètres, quantization 4-bit → 3GB VRAM
- Spécialisé imagerie médicale (chest X-ray, pathologie)
- Accès via HuggingFace (token requis + acceptation des conditions)

> ⚠️ Le GPU T4 (14GB) ne supporte pas YOLO + LLM simultanément. YOLO est libéré avant le chargement de MedGemma.

### 🛠️ Définition des fonctions utilitaires (détection, annotation, prompt, LLM)

```python
import cv2, gc, os, random, torch
from PIL import Image
import matplotlib.pyplot as plt
import matplotlib.patches as patches

CLASS_NAMES = [
    'Aortic_enlargement', 'Atelectasis', 'Calcification',
    'Cardiomegaly', 'Consolidation', 'ILD', 'Infiltration',
    'Lung_Opacity', 'Nodule/Mass', 'Other_lesion',
    'Pleural_effusion', 'Pleural_thickening', 'Pneumothorax',
    'Pulmonary_fibrosis', 'No finding'
]
CONFIDENCE_THRESHOLDS = {'high': 0.70, 'medium': 0.45, 'low': 0.25}

def get_confidence_level(conf):
    if conf >= CONFIDENCE_THRESHOLDS['high']:
        return "HAUTE", "🟢", "Le modèle est très sûr"
    elif conf >= CONFIDENCE_THRESHOLDS['medium']:
        return "MOYENNE", "🟡", "Confiance modérée"
    else:
        return "FAIBLE", "🔴", "Incertain, interpréter avec précaution"

def run_yolo_detection(image_path, conf_threshold=0.25):
    results = yolo_model.predict(source=image_path, conf=conf_threshold, imgsz=640, device=0, verbose=False)
    detections = []
    for result in results:
        for box in result.boxes:
            cls_id = int(box.cls[0])
            conf   = float(box.conf[0])
            x1, y1, x2, y2 = box.xyxy[0].tolist()
            conf_level, emoji, warning = get_confidence_level(conf)
            detections.append({
                'class_id': cls_id, 'class_name': CLASS_NAMES[cls_id],
                'confidence': conf, 'conf_level': conf_level,
                'emoji': emoji, 'warning': warning, 'bbox': [x1, y1, x2, y2]
            })
    detections.sort(key=lambda x: x['confidence'], reverse=True)
    return detections

def draw_annotated_image(image_path, detections):
    img = cv2.imread(image_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    fig, ax = plt.subplots(1, 1, figsize=(12, 12))
    ax.imshow(img)
    COLORS = {'HAUTE': '#00C853', 'MOYENNE': '#FFD600', 'FAIBLE': '#FF1744'}
    for det in detections:
        if det['class_name'] == 'No finding':
            continue
        x1, y1, x2, y2 = det['bbox']
        color = COLORS[det['conf_level']]
        ax.add_patch(patches.Rectangle((x1, y1), x2-x1, y2-y1, linewidth=2, edgecolor=color, facecolor='none'))
        ax.text(x1, y1-5, f"{det['emoji']} {det['class_name']} {det['confidence']:.0%}",
                color='white', fontsize=8, fontweight='bold',
                bbox=dict(facecolor=color, alpha=0.8, pad=2, edgecolor='none'))
    ax.axis('off')
    ax.set_title('Détections YOLOv8 — Radiographie thoracique', fontsize=13, fontweight='bold')
    plt.tight_layout()
    annotated_path = '/content/annotated_xray.png'
    plt.savefig(annotated_path, dpi=150, bbox_inches='tight')
    plt.show()
    return annotated_path

def build_llm_prompt(detections):
    high_conf   = [d for d in detections if d['conf_level'] == 'HAUTE'   and d['class_name'] != 'No finding']
    medium_conf = [d for d in detections if d['conf_level'] == 'MOYENNE' and d['class_name'] != 'No finding']
    low_conf    = [d for d in detections if d['conf_level'] == 'FAIBLE'  and d['class_name'] != 'No finding']
    no_finding  = [d for d in detections if d['class_name'] == 'No finding']

    prompt = """You are an expert radiologist assistant analyzing a chest X-ray.
A YOLO detection model has identified the following findings:\n\n=== DETECTIONS ===\n"""

    if high_conf:
        prompt += "\n🟢 HIGH CONFIDENCE (model is very certain):\n"
        for d in high_conf:
            prompt += f"  - {d['class_name']} (confidence: {d['confidence']:.1%})\n"
    if medium_conf:
        prompt += "\n🟡 MEDIUM CONFIDENCE (moderate caution):\n"
        for d in medium_conf:
            prompt += f"  - {d['class_name']} (confidence: {d['confidence']:.1%})\n"
    if low_conf:
        prompt += "\n🔴 LOW CONFIDENCE (high caution):\n"
        for d in low_conf:
            prompt += f"  - {d['class_name']} (confidence: {d['confidence']:.1%})\n"
    if no_finding:
        prompt += f"\n✅ No pathological finding (confidence: {no_finding[0]['confidence']:.1%})\n"
    if not any([high_conf, medium_conf, low_conf, no_finding]):
        prompt += "\nNo findings detected.\n"

    prompt += """\n=== YOUR TASK ===
1. SUMMARY: Brief clinical summary of findings
2. HIGH PRIORITY: Detail high confidence findings and their clinical significance
3. UNCERTAIN: Comment on medium/low confidence detections
4. RECOMMENDATIONS: Suggest follow-up or additional imaging if needed
5. DISCLAIMER: Remind this is AI-assisted and must be confirmed by a radiologist
"""
    return prompt

def run_llm_analysis(annotated_path, prompt):
    pil_image = Image.open(annotated_path).convert('RGB')
    formatted_prompt = f"[INST] <image>\n{prompt} [/INST]"
    inputs = processor(
        text=formatted_prompt, images=pil_image, return_tensors="pt"
    ).to(llava_model.device)
    with torch.no_grad():
        output = llava_model.generate(
            **inputs,
            max_new_tokens=800, temperature=0.3,
            do_sample=True, repetition_penalty=1.1,
            pad_token_id=processor.tokenizer.eos_token_id
        )
    input_len = inputs['input_ids'].shape[1]
    return processor.decode(output[0][input_len:], skip_special_tokens=True).strip()

print("✅ Toutes les fonctions sont prêtes !")
```

### 🚀 Pipeline complète — Détection YOLO + Rapport MedGemma

Cette cellule exécute l'intégralité du pipeline en séquence :
1. Charge YOLO et sélectionne une image (aléatoire ou uploadée)
2. Détecte les anomalies et génère l'image annotée avec les indices de confiance
3. Libère YOLO de la VRAM
4. Charge MedGemma-4B
5. Construit le prompt structuré et génère le rapport clinique final

```python
from ultralytics import YOLO
from transformers import AutoProcessor, PaliGemmaForConditionalGeneration, BitsAndBytesConfig
from huggingface_hub import login
from google.colab import userdata
from PIL import Image
import gc, torch, os, random


# ── ÉTAPE 2 : Charge YOLO ─────────────────────────────────
print("\n📦 Chargement YOLO...")
MODEL_PATH = '/content/drive/MyDrive/vinbigdata_runs/vinbigdata_yolov8m_50ep/weights/best.pt'
yolo_model = YOLO(MODEL_PATH)
print("✅ YOLO chargé !")

# ── ÉTAPE 3 : Sélectionne une image ───────────────────────
import os, gc, torch
from google.colab import files
print("Comment veux-tu choisir l'image ?")
print("1 - Image aléatoire du test set")
print("2 - Uploader ta propre image")
choix = input("Ton choix (1 ou 2) : ").strip()

if choix == "2":
    print("\nSélectionne une radiographie depuis ton PC...")
    uploaded   = files.upload()
    image_path = list(uploaded.keys())[0]
    print(f"✅ Image uploadée : {image_path}")
else:
    import random
    test_img_dir = '/content/vinbigdata-yolo-subset/test/images'
    image_path   = os.path.join(test_img_dir, random.choice(os.listdir(test_img_dir)))
    print(f"✅ Image aléatoire : {image_path}")
# ── ÉTAPE 4 : Détection YOLO ──────────────────────────────
print("\n🔍 Détection YOLO en cours...")
detections     = run_yolo_detection(image_path, conf_threshold=0.25)
annotated_path = draw_annotated_image(image_path, detections)

print(f"\n✅ {len(detections)} détection(s) :")
for d in detections:
    print(f"   {d['emoji']} {d['class_name']:25s} — {d['confidence']:.1%} ({d['conf_level']})")

# ── ÉTAPE 5 : Libère YOLO ─────────────────────────────────
print("\n🧹 Libération YOLO...")
del yolo_model
gc.collect()
torch.cuda.empty_cache()
torch.cuda.reset_peak_memory_stats()
vram_free = torch.cuda.mem_get_info()[0] / 1024**3
print(f"✅ YOLO libéré — VRAM libre : {vram_free:.1f} GB")

# ── ÉTAPE 6 : Charge MedGemma ─────────────────────────────
import gc, torch
from transformers import AutoProcessor, AutoModelForImageTextToText, BitsAndBytesConfig
from PIL import Image

# Libère l'ancien modèle
try:
    del llava_model, processor
    gc.collect()
    torch.cuda.empty_cache()
    print("🧹 Ancien modèle libéré")
except: pass

bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16  # ✅ bfloat16 recommandé pour Gemma3
)

processor = AutoProcessor.from_pretrained("google/medgemma-4b-it")
llava_model = AutoModelForImageTextToText.from_pretrained(
    "google/medgemma-4b-it",
    quantization_config = bnb_config,
    device_map          = "auto",
    dtype               = torch.bfloat16,  # ✅ dtype au lieu de torch_dtype
    low_cpu_mem_usage   = True,
)
llava_model.eval()
vram_used = torch.cuda.memory_allocated() / 1024**3
print(f"✅ MedGemma rechargé — VRAM utilisée : {vram_used:.1f} GB")

def run_llm_analysis(annotated_path, prompt):
    pil_image = Image.open(annotated_path).convert('RGB')

    messages = [
        {
            "role": "user",
            "content": [
                {"type": "image", "image": pil_image},
                {"type": "text",  "text": prompt}
            ]
        }
    ]

    inputs = processor.apply_chat_template(
        messages,
        add_generation_prompt = True,
        tokenize              = True,
        return_dict           = True,
        return_tensors        = "pt"
    )
    inputs = {k: v.to(llava_model.device) for k, v in inputs.items()}
    input_len = inputs['input_ids'].shape[1]

    with torch.no_grad():
        output = llava_model.generate(
            **inputs,
            max_new_tokens        = 600,
            do_sample             = False,
            pad_token_id          = processor.tokenizer.pad_token_id,
            eos_token_id          = processor.tokenizer.eos_token_id,
        )

    generated = output[0][input_len:]
    print(f"Tokens générés : {len(generated)}")
    response = processor.tokenizer.decode(generated, skip_special_tokens=True).strip()
    return response

# Relance
print("\n🤖 Analyse MedGemma en cours...")
prompt   = build_llm_prompt(detections)
response = run_llm_analysis(annotated_path, prompt)

print("\n" + "="*60)
print("RAPPORT D'ANALYSE RADIOLOGIQUE")
print("="*60)
print(response if response else "Reponse vide")
print("="*60)
print("\nAVERTISSEMENT : Cette analyse est generee par IA.")
print("Elle ne remplace pas l'avis d'un radiologue qualifie.")
```
